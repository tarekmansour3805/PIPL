```python
import pandas as pd
import numpy as np
```

```python
xls = pd.ExcelFile('DataSet7734_USA_Consumers_test-ver2.xls')
df1 = pd.read_excel(xls, 'NewSource-07734ZipCode')
df2 = pd.read_excel(xls, 'PiplCurrent-07734ZipCode')
```

```python
#existing data for each column total number of enteries are 3587:
df1.info()
```

```python
# Filtering zip codes for only 7734:
df1=df1[df1['Zip']== 7734]
```

```python
# no Data dropped from the original table:
df1.info()
```

```python
#gathering address Information in one column:
df1['address'] = df1['Address'].str.cat(df1['City'],sep=" ")
df2['address'] = df2['street name'].str.cat(df2['city'],sep=" ")
```

```python
# drop any columns with empty address cell or Nan's:
#this also Shows the data that is needed if using address as the primary column to check validity of data or missing data
df1 = df1[pd.notnull(df1['address'])]
df2 = df2[pd.notnull(df2['address'])]
df2.info()
#Checking if there is any empty rows in the data
df1.dropna(how='all')
df2.dropna(how='all')
#Droping duplicate values using email as key
#haven't done this for PIPL data Sheet since it should already have no duplicate values
df1 = df1.drop_duplicates(subset='EMail', keep='first')
```

```python
#Checking for duplicate values between both sheets only showing none duplicate values in the NewSource sheet
ds1 = set(tuple(line) for line in df1.values)
ds2 = set(tuple(line) for line in df2.values)
df = pd.DataFrame(list(ds1.difference(ds2)), columns=df1.columns)
```

```python
# the below means that only 2084 enteries are new valid enteries based on the data per row,
#any data rows that are common have been dropped from the table.
# total number of data that we can buy is 2084 rows, out of the original 3587 enteries
df.info()
```
